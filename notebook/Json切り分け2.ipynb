```python
import numpy as np
import pandas as pd
import json
import ast


pd.set_option('display.max_columns', 1000)
pd.set_option('display.max_rows', 1000)

train = pd.read_csv('../data/input/train.csv')
test = pd.read_csv('../data/input/test.csv')

train.head(3)
```

### production_countries

```python
train_gen = train.loc[:, 'production_countries']
test_gen = test.loc[:, 'production_countries']

#学習データ切り分け
train_gen = train_gen.to_json()
train_gen_dict = json.loads(train_gen)
iso_train_dict = {}
name_train_dict = {}

for k, v in train_gen_dict.items() :
    if v is not None :
        k = int(k)
        v = ast.literal_eval(v)
        
        if len(v) >= 2 :
            iso_train_list = []
            name_train_list = []
            for i in range(len(v)) :
                iso_train_list.append(str(v[i]['iso_3166_1']))
                name_train_list.append(str(v[i]['name']))
                
            iso_train_dict[k] = ', '.join(iso_train_list)
            name_train_dict[k] = ', '.join(name_train_list)

        else:
            iso_train_dict[k] = v[0]['iso_3166_1']
            name_train_dict[k] = v[0]['name']
            
iso_train_se = pd.Series(iso_train_dict, iso_train_dict.keys(), name='production_countries_iso_3166_1')
name_train_se = pd.Series(name_train_dict, name_train_dict.keys(), name='production_countries_name')


#テストデータ切り分け
test_gen = test_gen.to_json()
test_gen_dict = json.loads(test_gen)
iso_test_dict = {}
name_test_dict = {}

for k, v in test_gen_dict.items() :
    if v is not None :
        k = int(k)
        v = ast.literal_eval(v)
        
        if len(v) >= 2 :
            iso_test_list = []
            name_test_list = []
            for i in range(len(v)) :
                iso_test_list.append(str(v[i]['iso_3166_1']))
                name_test_list.append(str(v[i]['name']))
                
            iso_test_dict[k] = ', '.join(iso_test_list)
            name_test_dict[k] = ', '.join(name_test_list)

        else:
            iso_test_dict[k] = v[0]['iso_3166_1']
            name_test_dict[k] = v[0]['name']
            
iso_test_se = pd.Series(iso_test_dict, iso_test_dict.keys(), name='production_countries_iso_3166_1')
name_test_se = pd.Series(name_test_dict, name_test_dict.keys(), name='production_countries_name')
```

```python
train = pd.concat([train, iso_train_se, name_train_se], axis=1)
train.drop('production_countries', inplace=True, axis=1)
train.head(3)
```

```python
train = pd.concat([test, iso_test_se, name_test_se], axis=1)
train.drop('production_countries', inplace=True, axis=1)
train.head(3)
```
