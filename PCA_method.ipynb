# メソッド

```python
def standalization(X):
    """
    標準化を行う
    ------------
    X: データ行列
    return : st_X 標準化データ
    """
    mue =  np.mean(X)
    std = np.std(X)# std =√(np.sum((X - mue)**2 )))/len(X)
    return (X - mue)/std
```

```python
def covariance(X):
    """
    共分散
    ----------
    X: データ行列
    return : 共分散
    """
    return  (X.T.dot(X))/X.shape[0]
```

# 主成分分析法を一つのメソッドにまとめた

```python
def PCA(X, X_label):
    """
    主成分分析メソッド
    ---------------------
    X:　データ
    return 結果リスト
    """
    #平均標準化
    train_data = standalization(X)
    df_s_train_data = pd.DataFrame(X, columns=X_label)
    
    #共分散
    S_train = covariance(df_s_train_data) #sigmoid
    
    #固有ベクトル、固有値を求める
    U, S, V = np.linalg.svd(S_train) #特異値分解：U= 固有値, V = 固有ベクトル
    
    #削減する次元数kの選択
    rmse = np.sum(S)/np.sum(S_train)#平均二乗誤差
    
    result = np.array([])
    for data in rmse:
        if data > 0.99:
            result = np.append(result, "OK")
        else:
            result = np.append(result, "Concider reduce." )

    df_result = pd.concat([pd.Series(rmse.get_values()), pd.Series(result)], axis=1)
    df_result.index= rmse.index
    df_result.columns=["rmse", "result"]
    
    return df_result
```
